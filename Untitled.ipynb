```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(7, 5))

# Loop over sample indices 1 through 6
Runs = ["B","C","D"]

j=0
for i in range(1, 4):
    h = results["get_nd_tnp_histograms_1_hlt"][f"sample/{i}"]["passing"].project("el_pt")
    h.plot1d(ax=ax, label=f"EGamma0 Run {Runs[j]}")
    j=j+1

j=0
for i in range(4, 7):
    h = results["get_nd_tnp_histograms_1_hlt"][f"sample/{i}"]["passing"].project("el_pt")
    h.plot1d(ax=ax, label=f"EGamma1 Run {Runs[j]}")
    j=j+1

ax.set_xlim(0, 100)
ax.legend(title="Samples")
ax.set_xlabel("Electron pT [GeV]")
ax.set_ylabel("Counts")
ax.set_title("Passing Electrons: HLT_Ele23_Ele12_CaloIdL_TrackIdL_IsoVL_DZ")

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

h_pass = results["get_nd_tnp_histograms_1_hlt"]["sample/1"]["passing"].project("el_pt")
h_fail = results["get_nd_tnp_histograms_1_hlt"]["sample/1"]["failing"].project("el_pt")

# Get bin contents and edges
counts_pass = h_pass.values()
counts_fail = h_fail.values()
edges = h_pass.axes[0].edges  # <-- corrected: no parentheses

# Compute efficiency safely
num = counts_pass
den = counts_pass + counts_fail
eff = np.divide(num, den, out=np.zeros_like(num, dtype=float), where=den > 0)

# Bin centers for plotting
bin_centers = 0.5 * (edges[:-1] + edges[1:])

# Plot
fig, ax = plt.subplots(figsize=(7, 5))
ax.step(bin_centers, eff, where="mid")
ax.set_xlim(0, 100)
ax.set_ylim(0, 1)
ax.set_xlabel("Electron $p_T$ [GeV]")
ax.set_ylabel("Efficiency (passing / total)")
ax.grid(True)
plt.show()
```
