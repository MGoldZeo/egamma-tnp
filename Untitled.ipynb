```python
import awkward as ak
import matplotlib.pyplot as plt
import numpy as np

results = output['get_nd_tnp_histograms_1_hlt']
var = "el_pt"
visible_floor = 5e-2  # minimum visible floor for uncertainties

# Define the runs corresponding to your samples
runs = ["B", "C", "D"]
markers = ["o", "s", "^"]  # one marker per run (expand if needed)

# Map runs to sample indices in your data
run_map = {
    "B": ["sample/1", "sample/4"],
    "C": ["sample/2", "sample/5"],
    "D": ["sample/3", "sample/6"],
}

fig, ax = plt.subplots(figsize=(10, 8))
plotted_any = False

for run, marker in zip(runs, markers):
    keys = run_map.get(run)
    if not keys:
        continue

    # Combine EGamma0 and EGamma1 for this run
    h_pass = results[keys[0]]["passing"].project(var) + results[keys[1]]["passing"].project(var)
    h_fail = results[keys[0]]["failing"].project(var) + results[keys[1]]["failing"].project(var)

    total_hist = h_pass + h_fail

    # Convert to numpy
    total_counts = np.asarray(total_hist.values()).ravel()
    pass_counts = np.asarray(h_pass.values()).ravel()

    # Efficiency with safe division
    with np.errstate(divide='ignore', invalid='ignore'):
        eff = np.where(total_counts > 0, pass_counts / total_counts, 0.0)

    # Binomial uncertainties where total > 0
    errors = np.zeros_like(eff)
    nz = total_counts > 0
    errors[nz] = np.sqrt(eff[nz] * (1 - eff[nz]) / total_counts[nz])

    # enforce minimum visible floor
    errors = np.maximum(errors, visible_floor)

    # asymmetric bounds clipped to [0,1]
    upper_err = np.minimum(1.0 - eff, errors)
    lower_err = np.minimum(eff, errors)
    asym_err = np.vstack([lower_err, upper_err])

    # get bin centers
    bin_edges = total_hist.axes[0].edges
    bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

    # Plot with error bars
    ax.errorbar(
        bin_centers, eff, yerr=asym_err, fmt=marker, markersize=6,
        elinewidth=1.3, capsize=3, label=f"Run2024{run}"
    )

    plotted_any = True

if not plotted_any:
    raise RuntimeError("No runs found in results.")

# Final plot settings
ax.set_xlabel(r"E $p_T$ [GeV]")
ax.set_ylabel("Efficiency")
ax.set_title("2023 Trigger Efficiency by Run (BD), EGamma0+EGamma1 combined")
ax.set_xlim(0, 130)
ax.set_ylim(0, 1.05)
ax.grid(True)
ax.legend(title="Dataset run")
plt.tight_layout()
plt.show()
```
